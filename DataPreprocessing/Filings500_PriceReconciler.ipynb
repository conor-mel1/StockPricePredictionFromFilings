```python
price_data = pd.read_sql_query("SELECT * FROM price_data_500",
    conn)
price_data
```

```python
company_date_list = []
for company, date in zip(subset_balsheets['Company'], subset_balsheets['Time']):
    company_date_list.append([company, date])
    
print(len(company_date_list))
```

```python
# This piece of code takes 20 minutes exactly to run
# Need to change price to the price at the last day of the month filings were released in
reconciled_prices_list = [] 
for company_date_pair in company_date_list:
    df = price_data.loc[(price_data['Company'] == company_date_pair[0]) &
                        (price_data['Time'] == company_date_pair[1] + relativedelta(months=1) - datetime.timedelta(days = 1))]
    if len(df) != 0:
        print(df['Close'])
        reconciled_prices_list.append(float(df['Close'].tolist()[0]))
    elif len(df) == 0:
        print('Initial Date not a trading day')
        df = price_data.loc[(price_data['Company'] == company_date_pair[0]) &
                            (price_data['Time'] == company_date_pair[1] + relativedelta(months=1) - datetime.timedelta(days = 2))]
        
        if len(df) != 0:
            print(df['Close'])
            reconciled_prices_list.append(float(df['Close'].tolist()[0]))
        elif len(df) == 0:
            print('Initial Date plus 1 not a trading day')
            df = price_data.loc[(price_data['Company'] == company_date_pair[0]) &
                            (price_data['Time'] ==  company_date_pair[1] + relativedelta(months=1) - datetime.timedelta(days = 3))]
            
            if len(df) != 0:
                print(df['Close'])
                reconciled_prices_list.append(float(df['Close'].tolist()[0]))
            elif len(df) == 0:
                print('Initial Date plus 2 not a trading day')
                df = price_data.loc[(price_data['Company'] == company_date_pair[0]) &
                                (price_data['Time'] == company_date_pair[1] + relativedelta(months=1) - datetime.timedelta(days = 4))]
                
                if len(df) != 0:
                    print(df['Close'])
                    reconciled_prices_list.append(float(df['Close'].tolist()[0]))
                elif len(df) == 0:
                    print('Initial Date plus 3 not a trading day')
                    df = price_data.loc[(price_data['Company'] == company_date_pair[0]) &
                                (price_data['Time'] == company_date_pair[1] + relativedelta(months=1) - datetime.timedelta(days = 5))]
                    
                    if len(df) != 0:
                        print(df['Close'])
                        reconciled_prices_list.append(float(df['Close'].tolist()[0]))
                    elif len(df) == 0:
                        print('Initial Date plus 4 not a trading day')
                        df = price_data.loc[(price_data['Company'] == company_date_pair[0]) &
                                (price_data['Time'] == company_date_pair[1] + relativedelta(months=1) - datetime.timedelta(days = 6))]
                        
                        if len(df) != 0:
                            print(df['Close'])
                            reconciled_prices_list.append(float(df['Close'].tolist()[0]))
                        elif len(df) == 0:
                            print('Initial Date minus 5 not a trading day, Impossible')
```
