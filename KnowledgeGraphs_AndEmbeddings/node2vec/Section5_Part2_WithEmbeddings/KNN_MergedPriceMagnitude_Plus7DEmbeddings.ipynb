```python
Embedding_data = pd.read_csv("Company_7DEmbeddings.csv")
Embedding_data
```

```python
MFMagnitude_data = pd.read_csv("DeltaMF_data.csv")
MFMagnitude_data = MFMagnitude_data.sort_values(['Company', 'Time'])
MFMagnitude_data
```

```python
x = MFMagnitude_data.index[MFMagnitude_data['Company'] == '0'].tolist()
print(x)
for i in x:
    MFMagnitude_data.at[i, 'Company'] = 'UNM'
    
MFMagnitude_data = MFMagnitude_data.sort_values(['Company', 'Time'])
MFMagnitude_data
```

```python
data = pd.merge(MFMagnitude_data, Embedding_data, how='outer', on= ['Company'], left_on=None, right_on=None,
left_index=False, right_index=False, sort=True)
data
```

```python
# Shuffling the data to create random arrangement of data instances
data = data.sample(frac=1).reset_index(drop=True)
data
```

```python
prices = data['Price_Change'] 
data = data.drop(['Company', 'Time', 'Price_Change'], axis = 1)

# Normalising the data
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler() 
scaled_values = scaler.fit_transform(data) 
data.loc[:,:] = scaled_values

data.insert(data.shape[1], 'Price_Change', prices)
data
```

```python
# Converting Data to Numpy Arrays
NpMatrix = data.to_numpy(dtype = None, copy = False)
X = NpMatrix[:,0:33] # Parameters
y = NpMatrix[:,33] # Classification
print("X:", X)
print("\n")
print("Type X:", type(X))
print("Length of Individual X_train Vector:", len(X[1]))
print("Total Number of Training instances:", len(X))
print("\n")
print(y)
print("\n")
print("Type y:", type(y))
print("Length of Individual y_train vector", y[0])
print("Total number of y values", len(y))
```

```python
# Grid Search for Optimal KNN Hyperparameters
# Using GridSearchCV to find best parameters
kvals = range(1,100)
params = {'n_neighbors': kvals,
        'weights' : ['uniform', 'distance'],
         'metric' : ['euclidean', 'manhattan']}

knn = KNeighborsRegressor()
#mse = make_scorer(mean_squared_error, greater_is_better= False)
mae = make_scorer(mean_absolute_error, greater_is_better= False)

gs_modelKNN = GridSearchCV(knn, params, scoring = mae,  cv=5)
grid_results = gs_modelKNN.fit(X , y)

# Best parameters, best test score using these parameters, optimal KNN Configuration
print("Best Parameters", gs_modelKNN.best_params_)
print("Best score on Test Data", gs_modelKNN.best_score_)
print("Optimal Configuration", gs_modelKNN.best_estimator_)
```

```python
# Summarize the results in a readable format
print("Best: {0}, using {1}".format(grid_results.cv_results_['mean_test_score'], grid_results.best_params_))
results_df = pd.DataFrame(grid_results.cv_results_)
results_df
```

```python
results_df = results_df.sort_values(['rank_test_score'])
results_df
```

```python
CV = KFold(n_splits=10, random_state=None, shuffle=False)
```

```python
# Applying k-Fold Cross Validation 
# https://stackoverflow.com/questions/24132237/scikit-learn-cross-validation-scoring-for-regression#:~:text=How%20can%20one%20use%20cross_val_score%20for%20regression%3F%20The,accuracy%2C%20which%20is%20not%20very%20meaningful%20for%20regression.
#mse = make_scorer(mean_squared_error, greater_is_better= False)
mae = make_scorer(mean_absolute_error, greater_is_better= False) 
scores = cross_val_score(knn, X, y, scoring= mae, cv = CV) 

print(scores)

print("Mean MAE:", scores.mean())
print("SD of the mean:", scores.std())
```
