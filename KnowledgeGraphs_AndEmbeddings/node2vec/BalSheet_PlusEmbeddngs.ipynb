```python
import pandas as pd
import numpy as np

BS_data = pd.read_csv("BS_data.csv")
BS_data = BS_data.sort_values(['Company', 'Time'])
BS_data
```

```python
Embedding_data = pd.read_csv("Company_Embeddings.csv")
Embedding_data
```

```python
Emb_BS_data = pd.merge(BS_data, Embedding_data, how='outer', on= ['Company'], left_on=None, right_on=None,
left_index=False, right_index=False, sort=True)
Emb_BS_data
```
